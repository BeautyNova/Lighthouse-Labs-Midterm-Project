```python
import openpyxl
import pandas as pd
import json
import re
import numpy as np
```

```python
#create dataframe
df = pd.read_excel("FoursquareResults1.xlsx")
df
```

```python
#remove inconsequential address labels
df = df[['fsq_id', 'categories', 'popularity', 'price', 'rating', 'location.neighborhood', 'location.locality', 'stats.total_photos', 'stats.total_ratings', 'stats.total_tips']]
df.head()
```

```python
#rename columns
df.rename(columns = {'location.neighborhood' : 'neighborhood', 'location.locality' : 'borough', 'stats.total_ratings' : 'nbr_ratings', 'stats.total_tips' : 'tips', 'stats.total_photos' : 'photos'}, inplace = True)
df.head()
```

```python
df2 = df.copy()
```

```python
#clean up 
df2['neighborhood'] = df2['neighborhood'].str.strip("['']")
df2.head()
```

```python
#drop null values
df2.dropna(subset = ['price', 'rating', 'neighborhood', 'borough', 'photos', 'nbr_ratings', 'tips'], inplace=True)
df2
```

```python
df3 = df2.copy()
counter = 0

for i in df3['categories']:
    #using regex to grab categories name
    s = df3.iloc[counter]['categories']
    pattern = "name': '(.*?)', 'icon"
    
    substring = re.findall(pattern, s)
    
    df3.iat[counter, df.columns.get_loc('categories')] = substring
    counter = counter + 1
```
